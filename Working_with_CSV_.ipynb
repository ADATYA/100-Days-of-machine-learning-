## Sep Parameter

Here is the code to download and load the MovieLens 100k dataset's movie metadata file, which is in a TSV format.

```python
import requests
from io import StringIO
import pandas as pd

url = "http://files.grouplens.org/datasets/movielens/ml-100k/u.item"
headers = {"User-Agent": "Mozilla/5.0 (Macintosh; Intel Mac OS X 10.14; rv:66.0) Gecko/20100101 Firefox/66.0"}
req = requests.get(url, headers=headers)
data = StringIO(req.text)

# The u.item file is tab-separated and has no header row
# The columns are: movie id | movie title | release date | video release date | IMDb URL | unknown |
#                  Action | Adventure | Animation | Children's | Comedy | Crime | Documentary | Drama |
#                  Fantasy | Film-Noir | Horror | Musical | Mystery | Romance | Sci-Fi | Thriller |
#                  War | Western
# We will use a subset of these columns and name them appropriately
column_names = ['movie_id', 'movie_title', 'release_date', 'video_release_date', 'IMDb_URL', 'unknown',
                'Action', 'Adventure', 'Animation', 'Children\'s', 'Comedy', 'Crime', 'Documentary', 'Drama',
                'Fantasy', 'Film-Noir', 'Horror', 'Musical', 'Mystery', 'Romance', 'Sci-Fi', 'Thriller',
                'War', 'Western']
df_movies = pd.read_csv(data, sep='|', encoding='latin-1', names=column_names)

# Display the first few rows and the columns we are interested in
display(df_movies[['movie_id', 'movie_title', 'release_date', 'IMDb_URL']].head())
```

## Skiprows and nrows parameter

## dtypes parameter

```python
df  = pd.read_csv("/content/UberDataset.csv", dtype={"START_DATE":str})
display(df.head())
```

## Handling Dates

```python
df = pd.read_csv("/content/UberDataset.csv", parse_dates=["START_DATE","END_DATE"])
df.info()
display(df.head())
```

## Convertors

```python
def rename (name):
  if name == "START":
    return "S"
  else:
    return "E"

df = pd.read_csv("/content/UberDataset.csv", converters={"START":rename})
display(df.head())
```

```python
df = pd.read_csv("/content/UberDataset.csv", converters={"START":str})
display(df.head())
```

## na_values parameter

```python
df= pd.read_csv("/content/UberDataset.csv", na_values=["START_DATE"])
display(df.head())
```

## Loading a huge dataset in chunks

```python
df= pd.read_csv("/content/UberDataset.csv", chunksize=1000)
for chunks in df:
  display(chunks.head())
```

```python
dfs = pd.read_csv("/content/UberDataset.csv", chunksize=1000)
for chunks in dfs:
  display(chunks.memory_usage())
```
