#### Train test and split

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.1)
```

```python
X_train
```

```python
y_train
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
```

```python
scaleer = StandardScaler()
X_train = X_train.drop('Name', axis=1)
X_test = X_test.drop('Name', axis=1)
X_train = scaleer.fit_transform(X_train)
X_test = scaleer.transform(X_test)
```

```python
X_train
```

```python
X_test = scaleer.transform(X_test)
X_test
```

### Trian the **model**

```python
from sklearn.linear_model import LinearRegression
```

```python
clf = LinearRegression()
```

```python
clf.fit (X_train,y_train)
```

```python
y_pred = clf.predict(X_test)
```

```python
y_test
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuricy_score = (y_test,y_pred)
```

```python
accuricy_score
```

```python
from mlxtend.plotting import plot_decision_regions
```

```python
plot_decision_regions(X_train, y_train.values, clf=clf, legend=2)
```

```python
plt.scatter(y_test, y_pred)
plt.xlabel("Actual Graduation Year")
plt.ylabel("Predicted Graduation Year")
plt.title("Actual vs. Predicted Graduation Year")
plt.show()
```
